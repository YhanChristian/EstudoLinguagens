```python
# Imports
import pandas as pd
import requests
from bs4 import BeautifulSoup 
from tabulate import tabulate
```

```python
# URL
res = requests.get("https://www.climatempo.com.br/previsao-do-tempo/15-dias/cidade/539/santoandre-sp")
```

```python
# Parser
soup = BeautifulSoup(res.content,'html5lib')
```

```python
# Extrai a tabela do código HTML
print(soup)
```

```python
list_item = soup.find_all('div', attrs={'class': '-gray _flex _margin-l-5'})
```

```python
print(list_item)
```

```python
dados_prev = [x.text for x in soup.find_all('span', attrs={'class': '_margin-r-15'})]
```

```python
#exibe dados de temperatura e umidade extraídos
print(dados_prev)
```
